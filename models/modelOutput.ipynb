```python
from ultralytics import YOLO
from google.colab import files
from IPython.display import Image, display
import shutil
import os

# Load the trained YOLO model
model = YOLO('/content/drive/MyDrive/best_bone_fracture_model.pt')  # Adjust the path to your trained model

# Remove the runs directory if it exists to start fresh
if os.path.exists('runs/detect'):
    shutil.rmtree('runs/detect')

# Upload and move files
uploaded = files.upload()
for file_name in uploaded.keys():
    shutil.move(file_name, f'/content/{file_name}')
    image_path = f'/content/{file_name}'

# Make predictions and save results
results = model.predict(source=image_path, save=True, conf=0.25, save_txt=False, save_conf=False)

# Get the output directory for predictions
output_dir = 'runs/detect/predict/'
output_files = os.listdir(output_dir)
print("Files in prediction directory:", output_files)

# Display the first predicted output image
output_image_path = os.path.join(output_dir, output_files[0])
display(Image(output_image_path))
```
